```python
resposta = requests.get (API_BRASIL_IO) #chamando uma função que guarda respostas derivadas de uma outra variavél
if resposta.ok: # garantindo que não tenho nenhum erro ao executar os comandos
    dados = resposta.json() # obter os dados utilizando o método de json. json é uma função que está atrelado 
# ao request que herda caracteristicas que lida com o formato dos dados
resultados = dados['results'] # ciração do comando que irá executar o laço
dados # exibindos os dados # 4 elementos # se pedir para imprimir o comprimento, retorna 4 = len(dados)
```

```python
#Verificando que algumas cidades não conseguem ser analisadas, pois os dados foram retirados da primeira página. 
#Criando loop, para que possa varrer todas as páginas

def casoscidade(cidade):
    resposta = requests.get(API_BRASIL_IO) #chamando uma função que guarda respostas derivadas de uma outra variavél
    if resposta.ok: # garantindo que não tenho nenhum erro ao executar os comandos
        pagina = resposta.json() # obter os dados das páginas
    else:
        print ('Erro ao requisitar dados')
        return
    while True: 
        dados = pagina['results'] 
        for dado in dados:
            if dado['city'] == cidade and dado['is_last]' == True: #verificando se is_last é True
                print('Dados de', cidade) #imprimindo os dados que me interessa.
                print('Casos confirmados:', dado['last_available_confirmed']) # imprimindo dados confirmados
                print('Mortes confirmadas:', dado['last_available_deaths']) #imprimindo dados confirmados
                return # colocado somente para não ficar rodando dados toda vida
        if pagina['next'] is None: # retorna se encontrar
            return 
        else: # caso não encontre
            resposta = requests.get(pagina['next']) 
            if resposta.ok: 
                pagina = resposta.json() 
            else:
            print ('Erro ao requisitar dados')
            return

#Utiliozando while, pois, os dados são mudados todos os dias no brasil.io, se eu colocar for, amanhã,
#teriamos erro de vizualização de dados
# Para não ficar rodando para sempre, utilizamos True juntamente ao while
# Após obter os resultados eu utilizo o laço for para obter um dado, dentro de todos os dados para ver se refere-se dentro
# de uma cidade
        
```
